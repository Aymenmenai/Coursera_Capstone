```python
msk = np.random.rand(len(df)) < 0.5
train = df[msk]
test = df[~msk]
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.asanyarray(train[['Score']])
train_y = np.asanyarray(train[['Percent']])

test_x = np.asanyarray(test[['Score']])
test_y = np.asanyarray(test[['Percent']])


poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

```python
plt.plot(df.Score,df.Percent,'bo')
XX = (np.arange(0.0, 10.0, 0.1))
yy = ((clf.intercept_[0]+ clf.coef_[0][1]*XX+ clf.coef_[0][2]*np.power(XX, 2)))
plt.plot(XX+1, yy-0.48, '-r' )
plt.axvline(5,label='Happier countries')
plt.xlabel("Happiness")
plt.ylabel("Suidice")
ax=plt.gca()
ax.set_ylim([-1,3])
ax.set_xlim([0.5,10])
```
