<b>Here I created a loop if the borough has the "Not assigned" it will delete the row and if not it will give me the values of latiture 
and longiture and insert it in LAT and LNG on the dataframe</b>

```python
for i in range(180) :
    if BR[i]=='Not assigned':
        df.drop(i,axis=0,inplace=True)
    else :
        adress='{} , Toronto, ON, Canada'.format(BR[i])
        geolocator = Nominatim(user_agent="ca_explorer")
        location=geolocator.geocode(adress)
        latitude=location.latitude
        longitude=location.longitude
        df['LAT'][i]="{}".format(latitude)
        df['LNG'][i]="{}".format(longitude)
df
```

<b>Now the K-Means and Clustering</b>

```python
df.drop('Borough',axis=1,inplace=True)
```

```python
df.drop('Postal Code',axis=1,inplace=True)
```

```python
df.set_index('Neighbourhood')
print('ok')
```

```python
df
```

```python
kclusters = 2
df_clustering = df.drop('Neighbourhood',1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_clustering)
kmeans.labels_[0:10] 
```

```python
df['Cluster Labels']=kmeans.labels_
df
```

<b>Now the last map</b>

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lon, poi, cluster in zip(df['LAT'], df['LNG'], df['Neighbourhood'], df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster],
        fill=True,
        fill_opacity=0.7).add_to(map_clusters)
map_clusters
```
